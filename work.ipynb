```python
import pandas as pd
```

```python
data = pd.read_csv("creditcard.csv")
```

```python
data
```

```python
data.iloc[:, list(range(0,30))]
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model = sm.Logit(data["Class"], data.iloc[:, list(range(0,30))])
```

```python
result = model.fit()
```

```python
result.summary()
```

```python
# Note that the coefficient for "Time" column is really low, and that is suspicious
result.params["Time"]
# It is clear that we will need to engineer this feature for better predictive power contribution
```

```python
# deselect a few columns as they seem to be statistically insignificant based on above summary when predicting class
cols = [col for col in data.columns if col not in ['V12', 'V24']]
data_f = data[cols]
```

```python
data_f.columns
```

```python
# now, we engineer the time column since it is currently in seconds
data_f["Days"] = data_f["Time"]/(60*60*24)
data_f2 = data_f.drop(["Time"], axis= 1)
```

```python
data_f2.columns
```

```python
model_2 = sm.Logit(data_f2["Class"], data_f2[data_f2.columns.difference(["Class"])])
result_2 = model_2.fit()
```

```python
result_2.summary()
```

```python
# Let's check if our little maneuver helped improve the predictive power of the time-related feature
result_2.params["Days"]
# Looks good!
```

```python
# Let's check if the summary stats indicate that we might need to engineer the amount feature as well
data_f2[["Days","Amount"]].describe()
# I think we do, let's proceed
```

```python
# Engineering the Amount feature so that it is a multiple of hundred units (whether that is Euros or Dollars)
data_f2["Charged_amount"] = data_f2["Amount"]/100
```

```python
data_f3 = data_f2[data_f2.columns.difference(["Amount"])]
data_f3
```

```python
model_3 = sm.Logit(data_f3["Class"], data_f3[data_f3.columns.difference(["Class"])])
result_3 = model_3.fit()
```

```python
result_3.summary()
# Looking good, the coefficient for transaction amount is better now.
```

```python
# Let's build a few visualizations to get a feel of our data
import seaborn as sns
```

```python
sns.scatterplot(x = "Charged_amount", y= "Class", data = data_f3)
```

```python
# We can see, that our dataset is quite skewed. We'll deal with this later, down the line
data_f3.Class.value_counts()
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
X = data_f3[data_f3.columns.difference(["Class"])]
y = data_f3["Class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### The splits seem to be fine for now, let's train our model using these to get a reference score

```python
model_4 = LogisticRegressionCV(n_jobs = -1)
model_4.fit(X_train, y_train)
```

```python
y_pred = model_4.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
matrix_model_4 = confusion_matrix(y_test, y_pred)
print(matrix_model_4)
```

```python
from sklearn.metrics import classification_report
report_model_4 = classification_report(y_test, y_pred)
print(report_model_4)
```

##### Looks like our reference model is giving us a precision of 88% and recall of 63%.

#### Just to be clear, 

The precision [tp / (tp + fp)] is intuitively the ability of the classifier to not label a sample as positive if it is negative. Thus a high precision score denotes that the model will have less false positives.

The recall [tp / (tp + fn)] is intuitively the ability of the classifier to find all the positive samples. The recall, thus, is our model's ability to detect positives amongst a set of observed positives.

```python
from sklearn.metrics import roc_auc_score, roc_curve
model_4_roc_auc = roc_auc_score(y_test, y_pred)
```

```python
# plot characteristics curve
import matplotlib.pyplot as plt 
fpr, tpr, thresholds = roc_curve(y_test, model_4.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % model_4_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

### Now that we have our reference (vanilla) model, we bring in improvements

```python
# Before we employ any resampling techniques to deal with our skewed dataset, let's try throwing a Random Forest
# classifier at this dataset
from sklearn.ensemble import RandomForestClassifier

model_5 = RandomForestClassifier(n_jobs = -1, class_weight = "balanced")
model_5.fit(X_train, y_train)
```

```python
y_pred = model_5.predict(X_test)
print(classification_report(y_test, y_pred))
```

##### The RF classifier is giving us a precision score of 96% and recall of 74%. It beats the Logistic Regression classifier hands down, so we'll be using RF classifier going forward.

```python
# plot characteristics curve
model_5_roc_auc = roc_auc_score(y_test, y_pred)
fpr, tpr, thresholds = roc_curve(y_test, model_5.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Random Forest Classifier (area = %0.2f)' % model_5_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('RF_ROC')
plt.show()
```

### Now that we can see the ROC curve, let's proceed to leveraging sampling techniques to see if it improves performance

First off, we proceed to oversampling, by creating synthetic samples from the minor class (fraud transactions), and then use k-nearest neighbors to create similar but randomly tweaked new observations.

```python
# Synthetic Minority Oversampling Technique
from imblearn.over_sampling import SMOTE
os = SMOTE(random_state = 0)
os_data_X, os_data_y = os.fit_sample(X_train, y_train)
```

```python
# Let's try throwing these data at our RF model now
model_6 = RandomForestClassifier(n_jobs = -1).fit(os_data_X, os_data_y)
```

```python
y_pred = model_6.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
```

Looks like we are getting an improved F-1 score of 0.86, which beats everything we've seen till now. Thus, the oversampling is clearly helping train our model better. Plus, we also see a marked boost in the Recall (82% now, versus 74% earlier)

```python
# plot characteristics curve
model_6_roc_auc = roc_auc_score(y_test, y_pred)
fpr, tpr, thresholds = roc_curve(y_test, model_6.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Random Forest Classifier (area = %0.2f)' % model_6_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('RF_oversamp_ROC')
plt.show()
```
